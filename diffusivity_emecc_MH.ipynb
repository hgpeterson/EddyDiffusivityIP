```python
import numpy as np
import matplotlib.pyplot as plt
from ebm import EBM

def getcov(x,lr,D):
    dx = 6371e3*np.abs(x.reshape((-1,1))-x.reshape((1,-1)))*np.pi/2
    try:
        cov = D.reshape((-1,1))@D.reshape((1,-1))*np.exp(-dx/lr)
    except:
        cov = D**2*np.exp(-dx/lr)
    return cov

# real data
data = np.load("data/h_Q_GFDL-CM3.npz")
h = data["h"]
x = data["x"]
Q = data["Q"]
hs = h[0]
hn = h[-1]
# spectral = False
# D = 2.6e-4*np.ones(len(x))
spectral = True
n_polys = 6
D = np.zeros(n_polys)
D[0] = 2.6e-4 
D[4] = -1e-4 

# solve
ebm = EBM(x, Q, D, hs, hn, spectral=spectral)
h_tilde = ebm.solve()
```

```python
def log_likelihood(theta, x, y, gamma_inv):
    model = EBM(x, Q, theta, hs, hn, spectral=spectral)
    diff = y-model.solve().reshape(y.shape)/1e3
    return -0.5 * (diff.reshape((1,-1))@gamma_inv@diff.reshape((-1,1))).flatten()[0]
```

```python
from scipy.optimize import minimize
gamma_inv = np.linalg.inv(getcov(x,1e6,5))
np.random.seed(42)
nll = lambda *args: -log_likelihood(*args)
initial = D
soln = minimize(nll, initial, args=(x, h/1e3, gamma_inv))
D_ml = soln.x
chat = soln.hess_inv
chat_inv = np.linalg.inv(chat)

print("Maximum likelihood estimates:")
for i in range(len(D_ml)):
    print(f'D{i} = {D_ml[i]:.3e}')
```

```python
plt.plot(x,h/1e3)
plt.plot(x,EBM(x, Q, D_ml, hs, hn, spectral=spectral).solve()/1e3)
```

```python
def log_prior(theta,mean,chat_inv):
    if np.all(theta > -1e-3) and np.all(theta < 1e-3):
        diff = mean-theta.reshape(mean.shape)
        return -0.5 * (diff.reshape((1,-1))@chat_inv@diff.reshape((-1,1))).flatten()[0]
    return -np.inf

def log_probability(theta, x, y, gamma_inv,chat_inv,mean):
    lp = log_prior(theta,mean,chat_inv)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, gamma_inv)
```

```python
import emcee

pos = soln.x + 1e-6 * np.random.randn(32, n_polys)
nwalkers, ndim = pos.shape
mean = soln.x #np.array([2.6e-4, 0, 0, 0, -1e-4, 0])
filename = "mhmove.h5"
backend = emcee.backends.HDFBackend(filename)
backend.reset(nwalkers, ndim)
sampler = emcee.EnsembleSampler(
    nwalkers, ndim, log_probability, args=(x, h/1e3, gamma_inv,chat_inv,mean),backend=backend,
    moves=[(emcee.moves.GaussianMove(chat),1),],
)
#sampler.run_mcmc(pos, 2000, progress=True);
```

```python
max_n = 10000

# We'll track how the average autocorrelation time estimate changes
index = 0
autocorr = np.empty(max_n)

# This will be useful to testing convergence
old_tau = np.inf

# Now we'll sample for up to max_n steps
for sample in sampler.sample(pos, iterations=max_n, progress=True):
    # Only check convergence every 100 steps
    if sampler.iteration % 100:
        continue

    # Compute the autocorrelation time so far
    # Using tol=0 means that we'll always get an estimate even
    # if it isn't trustworthy
    tau = sampler.get_autocorr_time(tol=0)
    autocorr[index] = np.mean(tau)
    index += 1

    # Check convergence
    converged = np.all(tau * 100 < sampler.iteration)
    converged &= np.all(np.abs(old_tau - tau) / tau < 0.01)
    if converged:
        break
    old_tau = tau
```

```python
import matplotlib.pyplot as plt

n = 100 * np.arange(1, index + 1)
y = autocorr[:index]
plt.plot(n, n / 100.0, "--k")
plt.plot(n, y)
plt.xlim(0, n.max())
plt.ylim(0, y.max() + 0.1 * (y.max() - y.min()))
plt.xlabel("number of steps")
plt.ylabel(r"mean $\hat{\tau}$");
```

```python
fig, axes = plt.subplots(6, figsize=(10, 15), sharex=True)
samples = sampler.get_chain()
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(f'c{i}')
    #ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```
